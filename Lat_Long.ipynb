```python
import pandas as pd
from matplotlib.pyplot import figure, show
import numpy as np
```

```python
data = pd.read_csv("latitude_longitude_details.csv")
```

```python
print(data)
```

```python
lat = data["latitude"].tolist()
long = data["longitude"].tolist()
mean_lat = np.mean(lat)
mean_long = np.mean(long)

print(mean_lat)
print(mean_long)
print(np.std(lat))
print(np.std(long))
```

```python
points = list(zip(lat, long)) 
```

```python
S = set(points)
F = {}
for i in list(S):
    F[i] =points.count(i)
```

```python
fig = figure()
sub = fig.add_subplot(111)
K = F.keys()
Xs = [i[0] for i in K]
Ys = [i[1] for i in K]

sub.plot(Xs, Ys, 'o')
for i in K:
    sub.annotate(F[i], xy=i)
    
show()
```
